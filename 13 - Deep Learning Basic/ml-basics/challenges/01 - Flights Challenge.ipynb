```python
import pandas as pd
import numpy
import stat as st
from matplotlib import pyplot as plt
```

```python
df_flights = pd.read_csv('data/flights.csv')
df_flights.head(2)
```

```python
# 1. 
# Identificar cualquier dato nulo o faltante e imputar valores de reemplazo apropiados.
df_flights.fillna(-1,inplace=True)
df_flights.DepDel15 = df_flights.DepDel15.astype('int32')

# Identifique y elimine cualquier valor atípico en las columnas **DepDelay** y **ArrDelay**.
porcentajes = [0, 0.01, 0.05, 0.25, 0.5, 0.75,0.9, 0.95, 1]

    # DepDelay Outliers
df_DepDelay = pd.DataFrame(df_flights['DepDelay'].quantile(porcentajes))
df_DepDelay.index = [f'{p * 100}%' for p in porcentajes]
print(df_DepDelay.transpose())

print('')   # Espacio
    # ArrDelay
df_ArrDelay = pd.DataFrame(df_flights['ArrDelay'].quantile(porcentajes))
df_ArrDelay.index = [f'{p * 100}%' for p in porcentajes]
print(df_ArrDelay.transpose())
```

```python
def outliers_graficas(var_data):
    '''Muestra de forma grafica la dispersion de datos:
        - Media
        - Mediana
        - Moda
        - Minimo
        - Maximo'''
    media = var_data.mean()
    mediana = var_data.median()
    moda = var_data.mode()[0]
    min = var_data.min()
    max = var_data.max()

    fig, ax = plt.subplots(2, 1, figsize = (10,4))
    # Histograma
    ax[0].hist(var_data)
    ax[0].set_ylabel('Frecuencia')
        # Lineas de demarcacion de datos 
    ax[0].axvline(x=min, color = 'gray', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=media, color = 'cyan', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=mediana, color = 'red', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=moda, color = 'yellow', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=max, color = 'gray', linestyle='dashed', linewidth = 2)
    
    # Diagrama de Cajas
    ax[1].boxplot(var_data,vert = False)
    ax[1].set_ylabel('Valores')

    fig.suptitle(var_data.name)

    plt.show()

```

```python
Campos_Delay = ['DepDelay','ArrDelay']

for colum in Campos_Delay:
    outliers_graficas(df_flights[colum])
```

```python


# Trim outliers for ArrDelay based on 1% and 90% percentiles
ArrDelay_01pcntile = df_flights.ArrDelay.quantile(0.01)
ArrDelay_90pcntile = df_flights.ArrDelay.quantile(0.90)
df_flights = df_flights[df_flights.ArrDelay < ArrDelay_90pcntile]
df_flights = df_flights[df_flights.ArrDelay > ArrDelay_01pcntile]

# Trim outliers for DepDelay based on 1% and 90% percentiles
DepDelay_01pcntile = df_flights.DepDelay.quantile(0.01)
DepDelay_90pcntile = df_flights.DepDelay.quantile(0.90)
df_flights = df_flights[df_flights.DepDelay < DepDelay_90pcntile]
df_flights = df_flights[df_flights.DepDelay > DepDelay_01pcntile]

# View the revised distributions
for col in Campos_Delay:
    outliers_graficas(df_flights[col])
```

```python
df_flights[Campos_Delay].describe()
```

```python

#- *¿Cuáles son los retrasos promedio (promedio) de salida y llegada?*

print(f'Retraso de Salida promedio: {df_flights.DepDelay.mean()} min')
print(f'Retraso de Llegada promedio: {df_flights.ArrDelay.mean()} min')
```

```python
#*¿Cómo se comparan los transportistas en términos de rendimiento de retrasos en la llegada?*
for colum in Campos_Delay:
    df_flights.boxplot(column=colum, by= 'Carrier', figsize=(8,8))
```

```python
# *¿Hay una diferencia notable en los retrasos de llegada para los diferentes días de la semana?*
for colum in Campos_Delay:
    df_flights.boxplot(column= colum, by = 'DayOfWeek', figsize=(8,8))
plt.show()
```

```python
df_flights.columns
```

```python

# *¿Qué aeropuerto de salida tiene el promedio de retraso de salida más alto?*
df = df_flights.groupby(df_flights.OriginAirportName)
df_mean_originairport = pd.DataFrame(df['DepDelay'].mean()).sort_values('DepDelay',ascending=False)
df_mean_originairport
df_mean_originairport.plot(kind="bar",figsize=(12,12))
```

```python

# *¿Las salidas **tardías** tienden a provocar retrasos de llegada más prolongados que las salidas puntuales?*
df_flights.boxplot(column='ArrDelay', by= 'DepDel15', figsize=(12,12))
plt.show()
```

```python
# *¿Qué ruta (del aeropuerto de origen al aeropuerto de destino) tiene más llegadas **retrasadas**?*
df_flights['Routes'] = df_flights.OriginAirportName + '>' + df_flights.DestAirportName

df_routes = df_flights.groupby(df_flights.Routes)
pd.DataFrame(df_routes.ArrDel15.sum()).sort_values('ArrDel15',ascending=False)
```

```python

# *¿Qué ruta tiene el promedio de retraso de llegada más alto?*
pd.DataFrame(df_routes.ArrDelay.mean()).sort_values('ArrDelay',ascending=False)
```
